### 1. Use Vectors to Plot these functions

- $y=-4x-5$
- $y=2x+16$
- $y=x^2-3x+2$
- $4x-5y=10$

```r
xx = 1:10
yy = -4*xx - 5
zz = -2*xx + 16
aa = xx^2 - 3*xx + 2
bb = .8*xx - 2

paste0("Functions")
cat("xx:", xx)
cat("\nyy:", yy)
cat("\nzz:", zz)
cat("\naa:", aa)
cat("\nbb:", bb)
```

```r
par(mfrow=c(2,2))
plot(xx,yy)
plot(xx,zz)
plot(xx,aa)
plot(xx,bb)
options(repr.plot.width=12,repr.plot.height=6)
abline(h=0);abline(v=0)
```

### 2. Systems of Equations

Plot the solution to these systems of equations:

```r
A1 = matrix(c(4,-2,1,1), ncol=2)
b1 = c(-5,16)
z1 = solve(A1,b1)
```

```r
A2 = matrix(c(-2,4,1,-5), ncol=2)
b2 = c(16,10)
z2 = solve(A2,b2)
```

```r
A3 = matrix(c(4,4,1,-5), ncol=2)
b3 = c(-5,10)
z3 = solve(A3,b3)
```

```r
par(mfrow=c(2,2))
plot(z1)
plot(z2)
plot(z3)
```

#### Plot these two points

Use linear algebra to fit a line to these two points.

We know that the equation looks like this

$$y_i=\beta_0+\beta_1x_i$$

Which we can rewrite as

$$(1,x_i)^T(\beta_0, \beta_1)=y_i$$

or

$$\left(\begin{matrix}1&x_i\end{matrix}\right)\left(\begin{matrix}\beta_0\\ \beta_1\end{matrix}\right)=y_i$$

We can use this to make

$$X\beta=\mathbf{y}$$
